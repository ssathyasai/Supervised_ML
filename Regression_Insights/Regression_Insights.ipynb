**Hypotheses:**  
- H0: distribution of **tip** for yellow taxi = distribution of **tip** for green taxi
- H1: distribution of **tip** for yellow taxi â‰  distribution of **tip** for green taxi

```python
yellow_tip = df['tip'].loc[df['color']=='yellow']
green_tip = df['tip'].loc[df['color']=='green']
```

```python
green_tip.shape
```

```python
yellow_tip.shape
```

```python
# Independent T-Test
from scipy import stats

t_stat, p_val = stats.ttest_ind(yellow_tip,green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python
#alpha=0.05 -- significance level
if p_val < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
from scipy import stats

credit_tip = df['tip'][df['payment']=='credit card']
cash_tip = df['tip'][df['payment']=='cash']

t_stat, p_val = stats.ttest_ind(credit_tip, cash_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: The tip for payment through credit card is different from payment through cash")
else:
    print("Accept null hypothesis: No significant difference by Payment method")
```
